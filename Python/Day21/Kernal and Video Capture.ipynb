# Video

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    cv2.imshow("webcam", frame)
    
    c=cv2.waitKey(1)
    
    if c== 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

# Blurred video

```python
cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    kernel_10 = np.ones((1,20), dtype=np.float32)/(20)
    reshaped=kernel_10.reshape(-1,1)
    
    output = cv2.filter2D(frame,-1,reshaped)
    
    cv2.imshow("webcam", output)
    
    c=cv2.waitKey(1)
    
    if c== 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

cap = cv2.VideoCapture(0)

i=0

kernel_10 = np.ones((1,10), dtype=np.float32)/(10)
reshaped=kernel_10.reshape(-1,1)
    
while True:
    _, frame = cap.read()
    
    output = cv2.filter2D(frame,-1,reshaped)
    
    cv2.imshow("webcam", output)
    
    c=cv2.waitKey(1)
    
    if c==43 :
        i=i+20
        kernel_10 = np.ones((1,10+i), dtype=np.float32)/(10+i)
        reshaped=kernel_10.reshape(-1,1)
        
    if c==45 and i>10:
        i=i-10
        kernel_10 = np.ones((1,i), dtype=np.float32)/i
        reshaped=kernel_10.reshape(-1,1)
        
    elif c== 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
