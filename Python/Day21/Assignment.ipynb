```python
import math
import matplotlib.pyplot as plt
import numpy as np
import cv2

cap = cv2.VideoCapture(0)

original_kernel = np.array([[0,0,0],[0,1,0],[0,0,0]])

choice=""
blur=0
emboss=0

while True:
    _, frame = cap.read()
    
    output = cv2.filter2D(frame, -1, original_kernel)
    
    cv2.imshow("webcam", output)
    c=cv2.waitKey(1)
    
    if c== 27:
        break
        
    elif c==ord('b'):
        
        blur_kernel = np.ones((1,10), dtype=np.float32)/(10)
        original_kernel=blur_kernel.reshape(-1,1)
        choice='b'
        
    elif c == 43:
        if choice == 'b':
            blur=blur+20
            blur_kernel = np.ones((1,10+blur), dtype=np.float32)/(10+blur)
            original_kernel=blur_kernel.reshape(-1,1)
            
        elif choice == 'e':
            emboss=emboss+10
            original_kernel = np.array([[-2*emboss,-1*emboss,0], [-1*emboss,1*emboss,1*emboss],[0,1*emboss,2*emboss]])
            
            
    elif c==45 :
        if choice == 'b'and blur>10:
            blur=blur-10
            blur_kernel = np.ones((1,blur), dtype=np.float32)/blur
            original_kernel=blur_kernel.reshape(-1,1)
            
        elif choice == 'e' and emboss>1:
            emboss=emboss-1
            original_kernel = np.array([[-2*emboss,-1*emboss,0], [-1*emboss,1*emboss,1*emboss],[0,1*emboss,2*emboss]])
         
        elif choice == 'e' and emboss==1:
            original_kernel = np.array([[0,0,0],[0,1,0],[0,0,0]])
            
        elif choice == 'b' and blur<=10:
            original_kernel = np.array([[0,0,0],[0,1,0],[0,0,0]])
            
            
    elif c==ord('e'):
        choice='e'
        original_kernel=np.array([[-2,-1,0], [-1,1,1],[0,1,2]])
        
    elif c == ord('s'):         
        cv2.imwrite("me.jpg", output) 
        
    
        
cap.release()
cv2.destroyAllWindows()
```

# sin wave video

```python
import cv2
import math
import matplotlib.pyplot as plt
import numpy as np


cap = cv2.VideoCapture(0)

while True:
    _, frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    rows, cols = gray.shape
    
    img_output = np.zeros(gray.shape)
    
    for i in range(rows):
        for j in range(cols):
        
            offset_x = int(80.0*math.sin(2*3.14*i/180))
            offset_y = 0

            if j+offset_x < cols:
                img_output[i,j] = gray[i,(j+offset_x)]

            else:
                img_output[i,j]=0
    
    cv2.imshow("webcam", img_output)
    
    c=cv2.waitKey(1)
    
    if c== 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
