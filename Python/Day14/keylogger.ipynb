# Mouse ##solve the error

```python
from pynput import mouse
with open("mouse_mov.txt","a+") as f:
    def on_move(x, y):
        print('Pointer moved to {0}'.format((x, y)))
        f.write('Pointer moved to {0}'.format((x, y)))
        f.write("\n\n")

    def on_click(x, y, button, pressed):
        print('{0} at {1}'.format('Pressed' if pressed else 'Released',(x, y)))
        f.write('{0} at {1}'.format('Pressed' if pressed else 'Released',(x, y)))
        f.write("\n\n")
        if not pressed:
            # Stop listener
            return False

    def on_scroll(x, y, dx, dy):
        print('Scrolled {0} at {1}'.format('down' if dy < 0 else 'up',(x, y)))
        f.write('Scrolled {0} at {1}'.format('down' if dy < 0 else 'up',(x, y)))
        f.write("\n\n")

    # Collect events until released
    with mouse.Listener(on_move=on_move,on_click=on_click,on_scroll=on_scroll) as listener:
        listener.join()

    # ...or, in a non-blocking fashion:
    listener = mouse.Listener(on_move=on_move,on_click=on_click,on_scroll=on_scroll)
    listener.start()
```

# Keyboard

```python
from pynput import keyboard

def on_press(key):
    try:
        print('alphanumeric key {0} pressed'.format(
            key.char))
        with open('keylogger.txt','a') as f:
            f.write('special key {0} pressed'.format(
            key)+"\n")
    except AttributeError:
        print('special key {0} pressed'.format(
            key))
        with open('keylogger.txt','a') as f:
            f.write('special key {0} pressed'.format(
            key)+"\n")

def on_release(key):
    print('{0} released'.format(
        key))
    with open('keylogger.txt','a') as f:
            f.write('special key {0} pressed'.format(
            key)+"\n")
    if key == keyboard.Key.esc:
        # Stop listener
        with open('keylogger.txt','a') as f:
            f.write('special key {0} pressed'.format(
            key)+"\n")
        return False

# Collect events until released
with keyboard.Listener(
        on_press=on_press,
        on_release=on_release) as listener:
    listener.join()

# ...or, in a non-blocking fashion:
listener = keyboard.Listener(
    on_press=on_press,
    on_release=on_release)
listener.start()
```
