# Celsius/Farenheit Converter

```python
from tkinter import *

class Backend(Frame):
    def __init__(self,parent,*args,**kwargs):
        super().__init__(parent,*args,**kwargs)
        self.root=parent
        
        self.root.title("Degree Converter")
        self.root.geometry("400x400")
        
        self.fm=Frame(self.root,bg="#74B9FF",width=900,height=800)
        self.fm.place(x=70,y=90)
        
        self.user_input=Entry(self.fm,width=40)
        self.user_input.pack()
        
        self.btn1=Button(self.fm,fg="#192A56",bg="white",text="Celsius",command= lambda: self.Celsius())
        self.btn1.pack(padx=30,pady=30)
        
        self.lbl=Label(self.fm,fg='blue',bg="#74B9FF")
        self.lbl.pack()
        
        self.btn2=Button(self.fm,fg="#192A56",bg="white",text="Farenheit",command= lambda: self.Fahrenheit())
        self.btn2.pack(padx=30,pady=30)
    
    def Celsius(self):
        value=self.user_input.get()
        value=(int(value))
        value=(value - 32) * (5/9)
        value=int(value)
        self.lbl["text"]=str(value)+" C"
    
    def Fahrenheit(self):
        value=self.user_input.get()
        value=(int(value))
        value=(value*(9/5)) + 32
        self.lbl["text"]=str(value)+" F"

        
        
class App():
    def __init__(self):
        root=Tk()
        root.title("MyApp")
        root.geometry("300x400+40+40")
        w=Backend(root)

        root.mainloop()     
object=App()

```

```python
import tkinter as tk
from tkinter import ttk

object=Backend()

LARGE_FONT = ("Verdana", 12)


class Main_Window(tk.Tk):
    
    def __init__(self, *args, **kwargs):
        
        tk.Tk.__init__(self, *args, **kwargs)
        tk.Tk.wm_title(self, "File Handling")
        container = tk.Frame(self)
        container.pack(side = "top", fill = "both", expand = True)
        container.grid_rowconfigure(0, weight=1)
        container.grid_columnconfigure(0, weight=1)
        
        self.frames = {}
        
        for F in (Admin_page, Home_Page, Insert_Page, All_User, Edit_User, Delete_User):
        
            frame = F(container, self)

            self.frames[F] = frame

            frame.grid(row = 0, column = 0, sticky = "nsew")
        
        self.show_frame(Admin_page)
        
    def show_frame(self, cont):
        
        frame = self.frames[cont]
        frame.tkraise()

        
        
class Admin_page(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self, parent)
        label = tk.Label(self,text="ADMIN LOGIN",fg="white",bg="#AE1438",padx=50,pady=15)
        label.pack()
#         label.pack(fill=BOTH)

        
        
        mail=tk.Entry(self,width=50,borderwidth=5)
        mail.pack(padx=10,pady=10)

        password=tk.Entry(self,width=50,borderwidth=5)
        password.pack()
        
        button1 = ttk.Button(self, text = "Submit", command = lambda: self.log_in(mail.get(),password.get()).pack()


    def log_in(self,mail,password):
        try:
            if mail=="a@gmail.com" and password=="abc":
                 register = ttk.Button(self, text = "Go to Home Page", 
                             command = lambda: controller.show_frame(Home_Page)).pack()
            
        except:
            label = tk.Label(self,text="Not an admin",padx=50,pady=15)
            label.pack()


class Home_Page(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self,parent)
        label = tk.Label(self, text="Home Page", font = LARGE_FONT)
        label.pack(pady = 10, padx = 10)
        
        register = ttk.Button(self, text = "Register", 
                            command = lambda: controller.show_frame(Insert_Page)).pack()
        
        view_user = ttk.Button(self, text = "View All User", 
                            command = lambda: controller.show_frame(All_User)).pack()
        
        edit = ttk.Button(self, text = "Edit User", 
                            command = lambda: controller.show_frame(Edit_User)).pack()
        
        delete = ttk.Button(self, text = "Delete User", 
                            command = lambda: controller.show_frame(Delete_User)).pack()
        
class Insert_Page(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self,parent)
        
        label = tk.Label(self, text="Insert the User Details", font = LARGE_FONT)
        label.pack(pady = 10, padx = 10)
    
        labeln = tk.Label(self, text="Name", font = LARGE_FONT)
        labeln.pack()
        name=tk.Entry(self,width=30,borderwidth=5,font=LARGE_FONT)
        name.pack(pady=10)
        
        labelm = tk.Label(self, text="Email", font = LARGE_FONT)
        labelm.pack()
        mail=tk.Entry(self,width=30,borderwidth=5,font=LARGE_FONT)
        mail.pack(pady=10)
        
        labelp = tk.Label(self, text="Password", font = LARGE_FONT)
        labelp.pack()
        passwd=tk.Entry(self,width=30,borderwidth=5,font=LARGE_FONT)
        passwd.pack(pady=10)
        
        global nam
        nam=name.get()
        global email
        email=mail.get()
        global passd
        passd=passwd.get()
        
        button1 = ttk.Button(self, text = "Submit",
                             command = lambda:self.send_data([name.get(),mail.get(),passwd.get()])).pack()
        
    def send_data(self,data):
        if object.register(data):
            lbl1=tk.Label(self,text="Registration Successful!",font=LARGE_FONT)
            lbl1.pack()
        else:
            lbl2=tk.Label(self,text="Registration Unsuccessful!",font=LARGE_FONT)
            lbl2.pack()
        
class All_User(tk.Frame):
    
    def __init__(self, parent, controller):
        tk.Frame.__init__(self,parent)
        label = tk.Label(self, text="All Users", font = LARGE_FONT)
        label.pack(pady = 10, padx = 10)
        all_users = object.view_data()
        for items in all_users:
            name = tk.Label(self, text=items[0], font = LARGE_FONT)
            name.pack()
            mail = tk.Label(self, text=items[1], font = LARGE_FONT)
            mail.pack()
            passwd = tk.Label(self, text=items[2], font = LARGE_FONT)
            passwd.pack()
            new_line=tk.Label(self,text="\n ", font = LARGE_FONT)
            new_line.pack()
 
 
class Edit_User(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self,parent)
        
        label = tk.Label(self, text="Edit User", font = LARGE_FONT)
        label.pack(pady = 10, padx = 10)
        button1 = ttk.Button(self, text = "Back to Home", 
                             command = lambda: controller.show_frame(Home_Page)).pack()
        global answer
        global value
        global new_value
        
        labelp = tk.Label(self, text="name/mail", font = LARGE_FONT)
        labelp.pack()
        answers=tk.Entry(self,width=30,borderwidth=5,font=LARGE_FONT)
        answers.pack(pady=10)
        answer=answers.get()
        
        labelv = tk.Label(self, text="Enter Name/Mail", font = LARGE_FONT)
        labelv.pack()
        values=tk.Entry(self,width=30,borderwidth=5,font=LARGE_FONT)
        values.pack(pady=10)
        value=values.get()
        
        labeln = tk.Label(self, text="Enter new Name/Mail", font = LARGE_FONT)
        labeln.pack()
        new_values=tk.Entry(self,width=30,borderwidth=5,font=LARGE_FONT)
        new_values.pack(pady=10)
        new_value=new_values.get()
        
        
        button1 = ttk.Button(self, text = "Submit", 
                             command = lambda:self.edit_func(answers.get(),values.get(),new_values.get())).pack()
        
    def edit_func(self,answer,value,new_value):
        try:
            if object.edit_data(answer,value,new_value):
                lbl1=tk.Label(self,text="Data editted",font=LARGE_FONT)
                lbl1.pack()
            else:
                lbl2=tk.Label(self,text="Error in editting data.",font=LARGE_FONT)
                lbl2.pack()
                
        except:
            lbl3=tk.Label(self,text="Invalid Input.",font=LARGE_FONT)
            lbl3.pack()
            
        

class Delete_User(tk.Frame):
    def __init__(self, parent, controller):
        tk.Frame.__init__(self,parent)
        label = tk.Label(self, text="Delete User", font = LARGE_FONT)
        label.pack(pady = 10, padx = 10)
        button1 = ttk.Button(self, text = "Back to Home",command = lambda: controller.show_frame(Home_Page)).pack()
        id_lbl = tk.Label(self, text="Enter Id ", font = LARGE_FONT)
        id_lbl.pack()
        id=tk.Entry(self,width=30,borderwidth=5,font=LARGE_FONT)
        id.pack(pady=10)
        global user_id
        user_id=id.get()
        button1 = ttk.Button(self, text = "Submit", 
                             command = lambda: self.delete_user(id.get())).pack()
    def delete_user(self,user_id):
        
        try:
            if object.delete_data(int(user_id)):
                lbl1=tk.Label(self,text="Data deleted",font=LARGE_FONT)
                lbl1.pack()
            else:
                lbl2=tk.Label(self,text="Error in deleting data.",font=LARGE_FONT)
                lbl2.pack()
        except:
            lbl2=tk.Label(self,text="Invalid Input.",font=LARGE_FONT)
            lbl2.pack()
        
app = Main_Window()
app.mainloop()
```
