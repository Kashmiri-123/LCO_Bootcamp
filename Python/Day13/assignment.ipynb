```python
import time

driver=webdriver.Chrome(executable_path="C:\\Users\\Kashmiri\\Documents\\chromedriver_win32 (1)\\chromedriver.exe")

driver.get('https://open.spotify.com/')

new_music_friday=driver.find_element_by_xpath('/html/body/div[2]/div/div[4]/div[3]/div[1]/div/div/section/div/section/div/div[3]/div/section/div/div[2]/div/div/div[4]')
new_music_friday.click()

time.sleep(3)
for i in range(1, 94):
    
    try:
        author='//*[@id="main"]/div/div[4]/div[3]/div[1]/div/div/div[2]/div/section/div/div/div[2]/section/ol/div['+str(i)+']/div/li/div[2]/div/div[2]/span[1]/span/span/a'
        author_path=driver.find_elements_by_xpath(author)
        
        xpath = '//*[@id="main"]/div/div[4]/div[3]/div[1]/div/div/div[2]/div/section/div/div/div[2]/section/ol/div[' + str(i) + ']/div/li/div[2]/div/div[1]'
        elem = driver.find_element_by_xpath(xpath)
        
        time='//*[@id="main"]/div/div[4]/div[3]/div[1]/div/div/div[2]/div/section/div/div/div[2]/section/ol/div['+str(i)+']/div/li/div[4]/div/span'
        time_path=driver.find_element_by_xpath(time)
        
        print(elem.text)
        for auth in author_path:
            print(auth.text,end="   ")
        print(time_path.text,end="   ")
        print("\n\n")
    except:
        author='//*[@id="main"]/div/div[4]/div[3]/div[1]/div/div/div[2]/div/section/div/div/div[2]/section/ol/div['+str(i)+']/li/div[2]/div/div[2]/span[1]/span/span/a'
        author_path=driver.find_elements_by_xpath(author)
        
        xpath = '//*[@id="main"]/div/div[4]/div[3]/div[1]/div/div/div[2]/div/section/div/div/div[2]/section/ol/div[' + str(i) + ']/li/div[2]/div/div[1]'
        elem = driver.find_element_by_xpath(xpath)
        
        time='//*[@id="main"]/div/div[4]/div[3]/div[1]/div/div/div[2]/div/section/div/div/div[2]/section/ol/div['+str(i)+']/li/div[4]/div/span'
        time_path=driver.find_element_by_xpath(time)
        print(elem.text)
        for auth in author_path:
            print(auth.text,end="   ")
        print(time_path.text,end="   ")
        print("\n\n")
        

driver.close()
```
